```python
import pandas as pd
```

```python
btc_hourly = pd.read_csv('Data_Coin_price/btc_hourly.csv')
eth_hourly = pd.read_csv('Data_Coin_price/eth_hourly.csv')
doge_hourly = pd.read_csv('Data_Coin_price/doge_hourly.csv')
```

```python
btc_daily = pd.read_csv('Data_Coin_price/btc_daily.csv')
eth_daily = pd.read_csv('Data_Coin_price/eth_daily.csv')
doge_daily = pd.read_csv('Data_Coin_price/doge_daily.csv')
```

Manipulate daily data
- Transform to datetime format
- Rename the columns

```python
btc_daily= btc_daily[(btc_daily['datetime']>='2021-01-01') & (btc_daily['datetime']<'2022-01-01')]
btc_daily['datetime'] = pd.to_datetime(btc_daily['datetime']).dt.date
btc_daily.rename(columns={"closePriceUsd": "BTCUSD_close", "highPriceUsd": "BTCUSD_high",
                         "lowPriceUsd": "BTCUSD_low", "openPriceUsd": "BTCUSD_open"}, inplace=True)
btc_daily.tail()
```

```python
eth_daily= eth_daily[(eth_daily['datetime']>='2021-01-01') & (eth_daily['datetime']<'2022-01-01')]
eth_daily['datetime'] = pd.to_datetime(eth_daily['datetime']).dt.date
eth_daily.rename(columns={"closePriceUsd": "ETHUSD_close", "highPriceUsd": "ETHUSD_high",
                         "lowPriceUsd": "ETHUSD_low", "openPriceUsd": "ETHUSD_open"}, inplace=True)
eth_daily.tail()
```

```python
doge_daily= doge_daily[(doge_daily['datetime']>='2021-01-01') & (doge_daily['datetime']<'2022-01-01')]
doge_daily['datetime'] = pd.to_datetime(doge_daily['datetime']).dt.date
doge_daily.rename(columns={"closePriceUsd": "DOGEUSD_close", "highPriceUsd": "DOGEUSD_high",
                         "lowPriceUsd": "DOGEUSD_low", "openPriceUsd": "DOGEUSD_open"}, inplace=True)
doge_daily.head()
```

```python
#Combine daily data
daily_prices = btc_daily.merge(eth_daily, on='datetime').merge(doge_daily, on='datetime')
```

```python
daily_prices = daily_prices.set_index('datetime')
```

```python
daily_prices.to_csv('Coin_price_cleaned_daily.csv')
```

```python
daily_prices
```

Manipulate hourly data
- Transform to datetime format and remove timezone information
- Rename the columns

```python
btc_hourly['datetime'] = pd.to_datetime(btc_hourly['datetime']).dt.tz_localize(None)
btc_hourly.rename(columns={"closePriceUsd": "BTCUSD_close", "highPriceUsd": "BTCUSD_high",
                         "lowPriceUsd": "BTCUSD_low", "openPriceUsd": "BTCUSD_open"}, inplace=True)
btc_hourly= btc_hourly[(btc_hourly['datetime']>='2021-01-01') & (btc_hourly['datetime']<'2022-01-01 00:00:00')]
btc_hourly.head()
```

```python
eth_hourly['datetime'] = pd.to_datetime(eth_hourly['datetime']).dt.tz_localize(None)
eth_hourly.rename(columns={"closePriceUsd": "ETHUSD_close", "highPriceUsd": "ETHUSD_high",
                         "lowPriceUsd": "ETHUSD_low", "openPriceUsd": "ETHUSD_open"}, inplace=True)
eth_hourly= eth_hourly[(eth_hourly['datetime']>='2021-01-01') & (eth_hourly['datetime']<'2022-01-01 00:00:00')]
eth_hourly.head()
```

```python
doge_hourly['datetime'] = pd.to_datetime(doge_hourly['datetime']).dt.tz_localize(None)
doge_hourly.rename(columns={"closePriceUsd": "DOGEUSD_close", "highPriceUsd": "DOGEUSD_high",
                         "lowPriceUsd": "DOGEUSD_low", "openPriceUsd": "DOGEUSD_open"}, inplace=True)
doge_hourly= doge_hourly[(doge_hourly['datetime']>='2021-01-01') & (doge_hourly['datetime']<'2022-01-01 00:00:00')]
doge_hourly.head()
```

```python
#Combine hourly data, it has missing data
hourly_prices = btc_hourly.merge(eth_hourly, on='datetime').merge(doge_hourly, on='datetime')
hourly_prices = hourly_prices.set_index('datetime')
```

```python
hourly_prices.info()
```

```python
hourly_prices.to_csv('Coin_price_cleaned_hourly.csv')
```
