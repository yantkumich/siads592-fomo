```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import all the files and put them into one dictionary
data_btc = {}
for i in range(147): 
    data_btc[i]=pd.read_csv('Data_Google_trends/btc/data_btcperiod'+str(i)+'.csv').drop('isPartial', axis=1)
```

```python
#Change the first dataframe
data_btc[0]['btc_adjusted'] = data_btc[0]['btc']
data_btc[0]['bitcoin_adjusted'] = data_btc[0]['bitcoin']
```

```python
#Change the subsequent dataframes
for i in range(1,147):
    data_btc[i]['btc_adjusted'] = data_btc[i]['btc']/data_btc[i].iloc[0][1]*data_btc[i-1].iloc[-1][3]
    data_btc[i]['bitcoin_adjusted'] = data_btc[i]['bitcoin']/data_btc[i].iloc[0][2]*data_btc[i-1].iloc[-1][4]
```

```python
#Combine the dataframe
df_btc = pd.DataFrame()
for i in range(147):
    df_btc = df_btc.append(data_btc[i])
```

```python
#Remove the unadjusted raw data
df_btc = df_btc[['date','btc_adjusted','bitcoin_adjusted']].set_index('date')
#df.head()
```

```python
#Do the same for ETH
data_eth = {}
for i in range(147): 
    data_eth[i]=pd.read_csv('Data_Google_trends/eth/data_ethperiod'+str(i)+'.csv').drop('isPartial', axis=1)
data_eth[0]['eth_adjusted'] = data_eth[0]['eth']
data_eth[0]['ethereum_adjusted'] = data_eth[0]['ethereum']
for i in range(1,147):
    data_eth[i]['eth_adjusted'] = data_eth[i]['eth']/data_eth[i].iloc[0][1]*data_eth[i-1].iloc[-1][3]
    data_eth[i]['ethereum_adjusted'] = data_eth[i]['ethereum']/data_eth[i].iloc[0][2]*data_eth[i-1].iloc[-1][4]
df_eth = pd.DataFrame()
for i in range(147):
    df_eth = df_eth.append(data_eth[i])
df_eth = df_eth[['date','eth_adjusted','ethereum_adjusted']].set_index('date')
```

```python
#Do the same for DOGE
data_doge = {}
for i in range(147): 
    data_doge[i]=pd.read_csv('Data_Google_trends/doge/data_dogeperiod'+str(i)+'.csv').drop('isPartial', axis=1)
    #replace 0 with 0.25
    data_doge[i]=data_doge[i].replace(0,0.25)
data_doge[0]['doge_adjusted'] = data_doge[0]['doge']
data_doge[0]['dogecoin_adjusted'] = data_doge[0]['dogecoin']

for i in range(1,147):
    data_doge[i]['doge_adjusted'] = data_doge[i]['doge']/data_doge[i].iloc[0][1]*data_doge[i-1].iloc[-1][3]
    data_doge[i]['dogecoin_adjusted'] = data_doge[i]['dogecoin']/data_doge[i].iloc[0][2]*data_doge[i-1].iloc[-1][4]
df_doge = pd.DataFrame()
for i in range(147):
    df_doge = df_doge.append(data_doge[i])
df_doge = df_doge[['date','doge_adjusted','dogecoin_adjusted']].set_index('date')
```

```python
df = df_btc.join(df_eth).join(df_doge)
#df.info()
```

```python
normalized_df=(df/df.max()*100).reset_index().drop_duplicates()
normalized_df=normalized_df[(normalized_df['date']>='2021-01-01') & (normalized_df['date']<='2022-01-01 00:00:00')]
normalized_df.rename(columns={"date": "datetime"}, inplace=True)
normalized_df = normalized_df.set_index('datetime')
```

```python
normalized_df.to_csv('Google_cleaned_hourly.csv')
```

```python
#Combining hourly data to get daily data
df_daily = df.reset_index()
df_daily.head()
```

```python
df_daily= df_daily[(df_daily['date']>='2021-01-01') & (df_daily['date']<'2022-01-01 00:00:00')].drop_duplicates()
df_daily['datetime'] = pd.to_datetime(df_daily['date']).dt.date
df_daily = df_daily.drop('date', axis=1)
df_daily = df_daily.groupby('datetime').sum()
normalized_df_daily=(df_daily/df_daily.max()*100)
```

```python
normalized_df_daily.to_csv('Google_cleaned_daily.csv')
```
