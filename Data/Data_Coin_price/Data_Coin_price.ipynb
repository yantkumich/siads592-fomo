```python
import san

dogecoin_1 = san.get("ohlc/dogecoin",
from_date="2020-01-01",
to_date="2021-01-01",
interval="1h")

dogecoin_2 = san.get("ohlc/dogecoin",
from_date="2021-01-02",
to_date="2022-01-01",
interval="1h")

dogecoin_3 = san.get("ohlc/dogecoin",
from_date="2022-01-02",
to_date="2022-04-09",
interval="1h")
```

```python
dogecoin = pd.concat([dogecoin_1, dogecoin_2, dogecoin_3], axis=0)
dogecoin
```
