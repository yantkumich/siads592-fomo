```python
from datetime import date
from datetime import timedelta
from pytrends.request import TrendReq
import pandas as pd
```

```python
start_date = date(2019,12,31)
periods_in_year = 61
year = 2
data_btc = {}
search_list = ['btc', 'bitcoin']

for period in range(year*periods_in_year):
    end_date = start_date + timedelta(days=5)
    start_year = start_date.year
    start_month = start_date.month
    start_day = start_date.day
    
    end_year = end_date.year
    end_month = end_date.month
    end_day = end_date.day
    
    
    
    pytrends = TrendReq(hl='en-US', tz=360)
    kw_list = search_list

    search_df = pytrends.get_historical_interest(kw_list, year_start=start_year,
                                             month_start=start_month, day_start=start_day,
                                             hour_start=23, year_end=end_year,
                                             month_end=end_month, day_end=end_day, hour_end=23,
                                             cat=0, geo='', gprop='', sleep=0)
    data_btc['period'+str(period)] = search_df
    start_date = start_date + timedelta(days=5)
```
