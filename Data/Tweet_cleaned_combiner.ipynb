```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_btc = pd.read_csv('Data_Tweet_counts/btc.csv')
df_eth = pd.read_csv('Data_Tweet_counts/eth.csv')
df_doge = pd.read_csv('Data_Tweet_counts/doge.csv')
```

```python
df_btc = df_btc[(df_btc['date']>='2021/01/01')&(df_btc['date']<'2022/01/01')]
df_btc = df_btc[['date', 'transactions', 'av_transaction_value', 'tweets', 'google_trends']]
df_btc['date'] = pd.to_datetime(df_btc['date'])
df_btc.rename(columns={"transactions": "BTC_transactions", "av_transaction_value": "BTC_av_trans_value_USD",
                         "tweets": "BTC_tweets", "google_trends": "BTC_google_trends", "date": "datetime"}, inplace=True)
df_btc = df_btc.set_index('datetime')
df_btc.head()
```

```python
df_eth = df_eth[(df_eth['date']>='2021/01/01')&(df_eth['date']<'2022/01/01')]
df_eth = df_eth[['date', 'transactions', 'av_transaction_value', 'tweets', 'google_trends']]
df_eth['date'] = pd.to_datetime(df_eth['date'])
df_eth.rename(columns={"transactions": "ETH_transactions", "av_transaction_value": "ETH_av_trans_value_USD",
                         "tweets": "ETH_tweets", "google_trends": "ETH_google_trends", "date": "datetime"}, inplace=True)
df_eth = df_eth.set_index('datetime')
df_eth.head()
```

```python
df_doge = df_doge[(df_doge['date']>='2021/01/01')&(df_doge['date']<'2022/01/01')]
df_doge = df_doge[['date', 'transactions', 'av_transaction_value', 'tweets', 'google_trends']]
df_doge['date'] = pd.to_datetime(df_doge['date'])
df_doge.rename(columns={"transactions": "DOGE_transactions", "av_transaction_value": "DOGE_av_trans_value_USD",
                         "tweets": "DOGE_tweets", "google_trends": "DOGE_google_trends", "date": "datetime"}, inplace=True)
df_doge = df_doge.set_index('datetime')
df_doge.head()
```

```python
df_tweets = df_btc.join(df_eth).join(df_doge)
```

```python
df_tweets.to_csv('Tweet_cleaned_daily.csv')
```
