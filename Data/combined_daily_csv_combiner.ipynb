```python
import pandas as pd
```

```python
#Hourly data
google_hourly = pd.read_csv('Google_cleaned_hourly.csv')
price_hourly = pd.read_csv('Coin_price_cleaned_hourly.csv')
social_hourly = pd.read_csv('Social_cleaned_hourly.csv')
```

```python
combined_hourly = google_hourly.merge(price_hourly, on ='datetime', how = 'left').merge(social_hourly, on='datetime', how='left')
```

```python
combined_hourly
```

```python
#Check missing value
#combined_hourly[combined_hourly.isnull().any(axis=1)]

#Fill NaN with prvious value (ffill)
combined_hourly = combined_hourly.fillna(method='ffill')
```

```python
combined_hourly.rename(columns={"btc_adjusted": "btc_google_adjusted", "bitcoin_adjusted": "bitcoin_google_adjusted",
                               "eth_adjusted": "eth_google_adjusted", "ethereum_adjusted": "ethereum_google_adjusted", 
                               "doge_adjusted": "doge_google_adjusted", "dogecoin_adjusted": "dogecoin_google_adjusted"}, 
                      inplace=True)
```

```python
combined_hourly.to_csv('Combined_data/combined_hourly.csv')
```

```python
#Daily data
google_daily = pd.read_csv('Google_cleaned_daily.csv')
price_daily = pd.read_csv('Coin_price_cleaned_daily.csv')
tweet_daily = pd.read_csv('Tweet_cleaned_daily.csv')
```

```python
combined_daily = google_daily.merge(price_daily, on ='datetime', how = 'left').merge(tweet_daily, on ='datetime', how = 'left')
```

```python
pd.set_option('display.max_columns', None)
combined_daily
```

```python
combined_daily.rename(columns={"btc_adjusted": "btc_google_adjusted", "bitcoin_adjusted": "bitcoin_google_adjusted",
                               "eth_adjusted": "eth_google_adjusted", "ethereum_adjusted": "ethereum_google_adjusted", 
                               "doge_adjusted": "doge_google_adjusted", "dogecoin_adjusted": "dogecoin_google_adjusted"}, 
                      inplace=True)
```

```python
combined_daily.to_csv('Combined_data/combined_daily.csv')
```
